#**DATA CLEANING**


```python
df.drop(columns=['zipcode','url','country_id'],inplace=True)
```

```python
df
```

```python
df['aggregate_rating']=df['aggregate_rating'].astype(float)
df['opentable_support']=df['opentable_support'].astype(float)
df['price_range']=df['price_range'].astype(int)
df['votes']=df['votes'].astype(int)
df
```

```python
df['cuisines'].fillna("Unknown",inplace=True)
df['timings'].fillna("Unknown",inplace=True)
df['address'].fillna("Not Available",inplace=True)
df['opentable_support'].fillna(0,inplace=True)
```

```python
df.describe()
```

```python
df.drop_duplicates(subset='res_id',keep='first',inplace=True)
df
```

```python
import ast # import the ast module
def convert_to_list(value):
  try:
    return ast.literal_eval(value) # safely covert string to list
  except(ValueError,SyntaxError):
      return[]
df['establishment']=df['establishment'].apply(convert_to_list)
df['highlights']=df['highlights'].apply(convert_to_list)
# Save cleaned dataset(optional)
df.to_csv("Cleaned_Indian_Restaurants.csv",index=False)
#Display dataset (optional)
print(df.info())
print(df.head())
```

#**EXPLORARTY** **DATA ANALYSIS (EDA)**

```python
top_brands = df["name"].value_counts().head(10)

# Plot the top 10 brands with the highest number of outlets
plt.figure(figsize=(12, 6))
sns.barplot(x=top_brands.values, y=top_brands.index, palette="coolwarm")
plt.xlabel("Number of Outlets")
plt.ylabel("Restaurant Brand")
plt.title("Top 10 Restaurant Brands with the Most Outlets")
plt.show()
```

```python
# @title TOP  10 CITIES WITH MOST RESTAURANTS
plt.figure(figsize=(12, 6))
top_cities = df['city'].value_counts().head(10)  # Get top 10 cities
sns.barplot(x=top_cities.values, y=top_cities.index, palette="coolwarm")
plt.xlabel("Number of Restaurants")
plt.ylabel("City")
plt.title("Top 10 Cities with Most Restaurants")
plt.show()
```

```python
# @title TOP MOST  POPULAR CUISINES IN INDIA
from collections import Counter

# Split multiple cuisines and count occurrences
cuisine_list = df['cuisines'].dropna().str.split(', ').sum()
cuisine_counts = Counter(cuisine_list)

# Convert to DataFrame and plot
cuisine_df = pd.DataFrame(cuisine_counts.items(), columns=['Cuisine', 'Count'])
cuisine_df = cuisine_df.sort_values(by='Count', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=cuisine_df['Count'], y=cuisine_df['Cuisine'], palette="viridis")
plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisine")
plt.title("Top 10 Most Popular Cuisines in India")
plt.show()
```

```python
# @title PRICE RANGE VS. RATINGS
plt.figure(figsize=(10, 6))
sns.boxplot(x="price_range", y="aggregate_rating", data=df, palette="Set2")
plt.xlabel("Price Range")
plt.ylabel("Aggregate Rating")
plt.title("Price Range vs. Ratings")
plt.show()
```

```python
# @title DISTRIBUTION OF RATINGS
plt.figure(figsize=(10, 6))
sns.histplot(df['aggregate_rating'], bins=20, kde=True, color='PURPLE')
plt.xlabel("Ratings")
plt.ylabel("Count")
plt.title("Distribution of Restaurant Ratings")
plt.show()
```

```python
# @title DELIEVERY VS. DINNING IN TRENDS
plt.figure(figsize=(8, 4))
delivery_counts = df['delivery'].value_counts()
labels = delivery_counts.index.tolist()
label_maping ={-1:'No Delivery',0: 'unknown',1: 'Delivery'}
labels =[label_maping.get(label,label) for label in labels]
plt.pie(delivery_counts, labels=labels, autopct='%1.1f%%', colors=['GREY','PINK','BLUE'])
plt.title("Percentage of Restaurants Offering Delivery")
plt.show()
```

```python
# @title TOP RATED RESTAURANTS IN INDIA
top_restaurants = df[df['aggregate_rating'] == 5][['name', 'city', 'cuisines', 'votes']]
print(top_restaurants.head(10))
```

```python
# @title VOTES VS. RATINGS
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["votes"], y=df["aggregate_rating"], alpha=0.5, color="PURPLE")

plt.xlabel("Number of Votes")
plt.ylabel("Aggregate Rating")
plt.title("Votes vs. Ratings")
plt.xlim(0, df["votes"].quantile(0.95))  # Limit extreme values for better visualization
plt.show()
```

```python
# @title DISTRIBUTION OF VOTES FOR RESTAURANTS
plt.figure(figsize=(10, 6))
sns.kdeplot(df["votes"], shade=True, color="BLUE")

plt.xlabel("Number of Votes")
plt.ylabel("Density")
plt.title("Distribution of Votes for Restaurants")
plt.xlim(0, df["votes"].quantile(0.95))  # Limit extreme values for better visualization
plt.show()
```

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(x=df["longitude"], y=df["latitude"], alpha=0.5, color="red")

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Geographical Distribution of Restaurants in India")
plt.show()
```

```python
# @title DENSITY OF PRICE RANGE
plt.figure(figsize=(10, 6))
sns.kdeplot(df["price_range"], shade=True, color="GREEN")

plt.xlabel("Price Range")
plt.ylabel("Density")
plt.title("Density of Price Ranges")
plt.show()
```

```python
# @title AVERAGE RATINGS BY PRICE RANGE
# Group by price range
price_rating = df.groupby("price_range")["aggregate_rating"].mean().sort_values()

# Plot bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=price_rating.index, y=price_rating.values, palette="Blues_r")

plt.xlabel("Price Range")
plt.ylabel("Average Rating")
plt.title("Average Ratings by Price Range")
plt.show()
```

```python
# @title NUMBER OF RESTAURANTS BY PRICE RANGE
plt.figure(figsize=(10, 6))
sns.countplot(x=df["price_range"], palette="magma")

plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.title("Number of Restaurants by Price Range")
plt.show()
```

# **CONCLUSION**

1️⃣ Major Cities Dominate the Restaurant Industry 🏙

.The highest number of restaurants are located in major metro cities like Delhi, Mumbai, Bangalore, and Hyderabad.

.This indicates that urban areas have a high demand for dining options, making them prime locations for new restaurants.

2️⃣ Indian Cuisine is the Most Popular, But Other Cuisines Are Growing 🍽

.The most frequently served cuisines include North Indian, Chinese, South Indian, and Fast Food.

.International cuisines like Italian, Continental, and Mexican are also gaining popularity, indicating a diverse food culture in India.

3️⃣ Higher Price Doesn’t Always Mean Higher Ratings 💰⭐

.The analysis showed that restaurants in higher price ranges do not always have the highest ratings.

.Mid-range restaurants (Price Range 2 & 3) tend to receive better ratings compared to the most expensive ones.

.This suggests that good food and service matter more than high prices for customer satisfaction.

4️⃣ Most Restaurants Have Moderate to High Ratings 📊

.The majority of restaurants have ratings between 3.5 to 4.5, showing that customers generally have a good dining experience.

.Very few restaurants received low ratings (<2.5), indicating that low-quality places struggle to survive in the competitive market.

5️⃣ Online Food Delivery is a Growing Trend 🚚

.A large percentage of restaurants now offer delivery services, reflecting the rise in online food ordering via Zomato, Swiggy, etc.

.This trend suggests that cloud kitchens and delivery-focused restaurants are becoming more profitable.

6️⃣ Bangalore Has the Most 5-Star Restaurants 🌟

.When analyzing top-rated restaurants (5-star ratings), Bangalore had the most
highly-rated places.

.This shows a strong preference for high-quality dining in tech hubs and metro cities.
